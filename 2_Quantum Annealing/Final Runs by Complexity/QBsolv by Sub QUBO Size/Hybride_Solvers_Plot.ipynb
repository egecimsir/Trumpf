## Hybride Solvers :  QBsolve as a Decomposer is highly affected by the structure of the underlying Problem

For two Job Sets -- which are very similar (both complexity 100) -- the results can vary drastically
```Jobs_100 A``` is our randomly generated Problem and ```Jobs_100 B```is a slightly simplified and manually constructed Problem

````python
import sys
sys.path.append("..")
from qubo_util import *
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv (r'data.csv', sep = ';')
````

````python
df 
````

````python
sns.set_theme(style="whitegrid")

# "Melt" the dataset to "long-form" or "tidy" representation
plot = pd.melt(df, "Job Set", var_name="Best Known Solutions (%)")

# Initialize the figure
f, ax = plt.subplots(figsize=(8,6))
sns.despine(bottom=True, left=True)


# Show the conditional means
sns.pointplot(x="Sub-QUBO Size", y="Best Known Solutions (%)", hue="Job Set",
              data=df,  
              #join=False, 
              linestyles=["--","--"],
              #palette="dark",
              #markers="d", 
              scale=0.45, 
              ci=None
             )

# Show the conditional means
sns.pointplot(x="Sub-QUBO Size", y="Best Known Solutions (%)", hue="Job Set",
              data=df, 
              join=False, 
              #palette="dark",
              markers="d", 
              scale=1.1, 
              ci=None
             )

# Improve the legend 
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[2:], labels[2:], title="Job Set",
          handletextpad=0, columnspacing=1,
          loc="lower left", ncol=3, frameon=True)

plt.savefig('Optimal Sub-QUBO Size.pdf')
````

While ```Jobs_100 A``` looses significantly BKS (%), ```Jobs_100 B```has no issues at all. QBSolv as a decomposer managed to structure the smaller Sub-Problems in ```Jobs_100 B``` way more efficiently

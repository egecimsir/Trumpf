```python
from qubo_util import *
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv (r'data.csv', sep = ';')
```

```python
df = df.drop(columns=['Average Energy', 'Standard Deviation', 'Lowest Energy'])
```

```python
df
```

```python
std = df.groupby(["Complexity", "QPU"], sort=False).std(ddof=0).reset_index()
```

```python
std
```

```python
sns.set_theme(style="whitegrid")

# "Melt" the dataset to "long-form" or "tidy" representation
plot = pd.melt(df, "QPU", var_name="Best Known Solutions (%)")

# Initialize the figure
f, ax = plt.subplots(figsize=(8,6))
sns.despine(bottom=True, left=True)

# Show each observation with a scatterplot
sns.stripplot(x="Complexity", y="Best Known Solutions (%)", hue="QPU",
              data=df, 
              dodge=True, 
              alpha=.35, 
              zorder=1
             )

# Show the conditional means
sns.pointplot(x="Complexity", y="Best Known Solutions (%)", hue="QPU",
              data=df, 
              dodge=.55, 
              #join=False, 
              linestyles=["--","--"],
              #palette="dark",
              #markers="d", 
              scale=0.45, 
              ci=None
             )

# Show the conditional means
sns.pointplot(x="Complexity", y="Best Known Solutions (%)", hue="QPU",
              data=df, 
              dodge=.55, 
              join=False, 
              #palette="dark",
              markers="d", 
              scale=1.1, 
              ci=None
             )

# Improve the legend 
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[4:], labels[4:], title="QPU",
          handletextpad=0, columnspacing=1,
          loc="upper right", ncol=3, frameon=True)

plt.savefig('Best Known Solutions by Complexity.pdf')
```
